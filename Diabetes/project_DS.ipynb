    Problem: There may be duplicate rows in the datasets.

#### 3. Remove NaN Values
    Problem: The datasets contain missing values.

#### 4. Remove Outliers
    Problem: The datasets may contain outliers.

#### 5. Remove Age After 10 Years
    Problem: The datasets may contain records of individuals older than 10 years.
    Solution: Filter out rows where age is greater than 10.

### Data Preprocessing

#### Data-1

```python
data1.fillna({
    "Age": data1["Age"].median(), 
    "BMI": data1["BMI"].mean(),  
    "Blood_Pressure": data1["Blood_Pressure"].mean(),
}, inplace=True)

data1.rename(columns=lambda x: x.strip().replace(" ", "_").lower(), inplace=True)


data1 = data1.astype({
    "age": "int32",
    "bmi": "float32",
    "blood_pressure": "float32",
})

data1 = data1[(data1["bmi"] >= 10) & (data1["bmi"] <= 50)]
```

#### Data-2

```python
data2.fillna({
    "Age": data2["Age"].median(),
    "bmi": data2["bmi"].mean(),
    "Blood Pressure": data2["Blood Pressure"].mean(),
}, inplace=True)

data2.rename(columns=lambda x: x.strip().replace(" ", "_").lower(), inplace=True)

data2 = data2.astype({
    "age": "int32",
    "bmi": "float32",
    "blood_pressure": "float32",
})

data2.rename(columns={'diabetes': 'diabetic'}, inplace=True)
data2 = data2[(data2["bmi"] >= 10) & (data2["bmi"] <= 50)]
```

```python
# Remove records where age is greater than 10
merged_data = pd.merge(data1, data2, on="id", how="inner")

# Remove columns with suffix _y
merged_data = merged_data.loc[:, ~merged_data.columns.str.endswith('_y')]

# Rename columns with suffix _x
merged_data.rename(columns=lambda col: col.replace("_x", ""), inplace=True)

id_col = merged_data.pop('id')  # Remove the 'ID' column
merged_data.insert(0, 'id', id_col)  # Insert it at the first position
```

### Clean of Merge_Data

```python
# Copying the dataset for cleaning
cleaned_data = merged_data.copy()

# Handling missing categorical values
cleaned_data['gender'].fillna('Unknown', inplace=True)
cleaned_data['smoking_status'].fillna('Unknown', inplace=True)

# Handling missing numerical values using median
numeric_cols = ['glucose_level', 'insulin_level', 'family_history', 'physical_activity', 
                'alcohol_consumption', 'diet_score', 'sleep_duration', 'occupation_stress_score', 'diabetic']

for col in numeric_cols:
    cleaned_data[col].fillna(cleaned_data[col].median(), inplace=True)

# Removing duplicates
cleaned_data.drop_duplicates(inplace=True)

# Removing outliers (e.g., BMI less than 10 or greater than 60)
cleaned_data = cleaned_data[(cleaned_data['bmi'] >= 10) & (cleaned_data['bmi'] <= 60)]

# Standardizing gender values to lowercase
cleaned_data['gender'] = cleaned_data['gender'].str.lower()

# Classifying gender values into 'male', 'female', and 'unknown'
valid_genders = ['male', 'female']
cleaned_data['gender'] = cleaned_data['gender'].apply(
    lambda x: x if x in valid_genders else 'unknown'
)

# Removing rows with 'unknown' gender
cleaned_data = cleaned_data[cleaned_data['gender'].str.lower() != 'unknown']

# Resetting indices after removal
cleaned_data.reset_index(drop=True, inplace=True)

```

#### Data Splitting

```python
file_path = r'clean_data1.csv'
data = pd.read_csv(file_path)

features = data.drop(columns='diabetic')
label = data['diabetic']
```

#### Encoding Categorical Features and Target Variable

```python
# Encode categorical features in X
label_encoder = LabelEncoder()
for column in features.select_dtypes(include=['object']).columns:
    features[column] = label_encoder.fit_transform(features[column].astype(str))

# Encode the target variable if it's categorical
if label.dtype == 'object':
    label = label_encoder.fit_transform(label)
```

### Split the data into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.25, random_state=42)
```

### Train a Decision Tree model


```python
# Encode categorical features in X_train and X_test
label_encoder = LabelEncoder()
for column in ['gender', 'smoking_status']:
	X_train[column] = label_encoder.fit_transform(X_train[column].astype(str))
	X_test[column] = label_encoder.transform(X_test[column].astype(str))

model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)
```

### Calculate model performance


```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted', zero_division=1)  
recall = recall_score(y_test, y_pred, average='weighted', zero_division=1)        
f1 = f1_score(y_test, y_pred, average='weighted', zero_division=1)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```

### Classification Report


```python
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
